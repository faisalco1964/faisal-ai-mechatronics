```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2');
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2',
               s=200,alpha=0.5);
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2',size ='salary');
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',size =400,style= 'level of education');
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
sns.set(style='darkgrid')
sns.displot(x='salary',data=df,bins=10,color='red',edgecolor='green',lw=2,kde=True,rug=True)
```

```python
sns.kdeplot(x='salary',data=df,fill=True);
```

```python
np.random.seed(42)
sample=np.random.randint(0,100,200)
sample
```

```python
sample =pd.DataFrame(sample,columns=['age'])
sample
```

```python
sns.rugplot(data=sample,x='age')
```

```python
sns.displot(x='age',data=sample,rug=True,kde=True,bins=15)
```

```python
sns.kdeplot(data=sample,x='age',clip=[0,100],bw_adjust=0.7,fill=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df
```

```python
df['division'].value_counts()
```

```python
sns.countplot(data=df,x='division',color='red',hue='level of education',palette='Set2')
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df,x='level of education',y ='salary',estimator=np.mean,errorbar='sd')
```
