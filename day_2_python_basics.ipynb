```python
#SET
sett = {'house','car','plane'}
sett.add('cloth')
print(sett)
sett2 = {'benz','toyota'}
sett.update(sett2)
print(sett)
sett.remove('house')
sett2.discard('toyota')
print(sett,sett2)
sett.pop()
print(sett)
sett.clear
print(sett)
#JOINING A SET
set1 = {2,5,6,3,35,}
set2 = {3,5,6,3,2,5,6,}
set3 = set1 | set2
set4 = set3.union(set2)
set5 = set1.intersect(set2)
```
