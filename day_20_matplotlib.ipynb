```python
 import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('my first plot');
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(0)
plt.ylim(0)
plt.title('my first plot');
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(0,8)
plt.ylim(0,16)
plt.title('my first plot')
plt.savefig('my first plt.png')
```
