```python
import numpy as np
m=np.linspace(0,10,11)
```

```python
m
```

```python
c = 3*10**8
```

```python
E = m*c**2
E
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('Mass')
ax.set_ylabel('Energy')
ax.set_title('E = MC^2')
ax.plot(E,m,color='red',lw=5)
```

```python
plt.plot(m,E,color='red',lw=5)
plt.title("E=mc^2")
plt.xlabel("Mass in Grams")
plt.ylabel("Energy in Joules")
plt.xlim(0,10)

# LOG SCALE
plt.yscale("log")
plt.grid(which='both',axis='y')

plt.show()
```
