```python
names =pd.Series(['andrew  ','bo;bo','  claire'  ,])
```

```python
names
```

```python
names.str.replace(';','').str.strip().str.capitalize()
```

```python
def cleenup(name):
    name = name.replace(';','')
    name =name.strip()
    name =name.capitalize()
    return name
```

```python
names.apply(cleenup)
```
