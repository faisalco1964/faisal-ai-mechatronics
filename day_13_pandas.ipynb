```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
```

```python
df
```

```python
def last_four(num):
    return str(num)[-4:]

df['last_four'] =df['CC Number'].apply(last_four)
df
```

```python
def func(num):
    if num <20:
        return 'good'
    elif 20<= num <=30:
        return 'better'
    else:
        return 'best'
func(30)
```

```python
df['emotion'] = df['total_bill'].apply(func)
df
```

```python
df['tip'].apply(lambda num: num*2)
```

```python
def quality(total_bill,tip):
    if tip/total_bill > 0.12:
        return 'generous'
    else:
        return 'other'
```

```python
df['Quality'] =df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis = 1)
```

```python
df
```

```python
def compare(size,Price_Per_Person):
    if Price_Per_Person/size > size:
        return 'high'
    else:
        return 'low'
```

```python
df[['size','price_per_person']].apply(lambda df:compare(df['size'],df['price_per_person']),axis = 1)
```
