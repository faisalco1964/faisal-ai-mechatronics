```python
df2=pd.merge(df1,sites,how ='inner',on='FILM')
```

```python
df2
```

```python
df2.info()
```

```python
df2['RT_NORM']=np.round(df2['RottenTomatoes']/20,1)
df2['RTU_NORM']=np.round(df2['RottenTomatoes_User']/20,1)
df2['MT_NORM']=np.round(df2['Metacritic']/20,1)
df2['MTU_NORM']=np.round(df2['Metacritic_User']/2,1)
df2['IBM_NORM']=np.round(df2['IMDB']/2,1)
```

```python
df2
```

```python
df2.columns
```

```python
norm=df2[['STARS','RATING','RT_NORM', 'RTU_NORM', 'MT_NORM',
       'MTU_NORM', 'IBM_NORM']]
```

```python
norm
```

```python
plt.figure(figsize=(15,6),dpi=500)
sns.kdeplot(data=norm,clip=(0,5),fill=True,palette='Set1')
```

```python
plt.figure(figsize=(15,6),dpi=500)
sns.kdeplot(data=norm[['STARS','RT_NORM']],clip=(0,5),fill=True,palette='Set1')
```

```python
plt.figure(figsize=(15,6))
sns.histplot(data=norm)
```

```python
plt.figure(figsize=(15,6))
sns.clustermap(data=norm,col_cluster=False)
```
