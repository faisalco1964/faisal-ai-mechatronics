```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
y=2*x
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
fig.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_ylabel('y axis')
axes[0][0].set_xlabel('x axis')
axes[0][0].set_title('fig1')
axes[0][1].plot(a,b)
axes[0][1].set_ylabel('y axis')
axes[0][1].set_xlabel('x axis')
axes[0][1].set_title('fig2')
axes[1][0].plot(x,y)
axes[1][0].set_ylabel('y axis')
axes[1][0].set_xlabel('x axis')
axes[1][0].set_title('fig3')
axes[1][1].plot(a,b)
axes[1][1].set_ylabel('y axis')
axes[1][1].set_xlabel('x axis')
axes[1][1].set_title('fig4')
fig.subplots_adjust(wspace=0.5,hspace=0.4)
fig.savefig('subplots',bbox_inches='tight')
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1)
for ax in axes:
    ax.plot(a,b)
fig.subplots_adjust(hspace=0.5)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1)
axes[0].plot(x,y)
axes[1].plot(a,b)
axes[2].plot(b,a)
fig.subplots_adjust(hspace=1)
```

```python
fig=plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',label='x vs y')
axes.plot(x,y+1,color='green',label='x vs y+1')
axes.plot(x,y+2,color='#e6c017',label='x vs y+2')


axes.legend(loc=(0.4,0.8))
```

```python
fig=plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',label='x vs y',lw=10)
axes.plot(x,y+1,color='green',label='x vs y+1',linewidth=6)#LINEWIDTH==LW
axes.plot(x,y+2,color='#e6c017',label='x vs y+2',lw=4)


axes.legend(loc=4)
```

```python
# possible linestype options --, , -., :, steps
fig, ax = plt.subplots(figsize=(6,6))

ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
ax.plot(x, x-2, color="green", lw=3, ls='-.') # dash and dot
ax.plot(x, x-3, color="green", lw=3, ls=':') # dots
ax.plot(x, x-4, color="green", lw=3, ls='--') # dashes
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
lines=axes.plot(x,y)
lines[0].set_dashes([1,5,2,2])
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,lw=1,color='red',marker='o')
axes.plot(x,y+1,lw=1,color='red',marker='+')
```
