```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
sns.jointplot(data=df,x='math score',y='reading score',color='green',hue='lunch')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',color='red',kind='hex')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',color='green',kind='kde',fill=True,hue='gender')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',color='purple',kind='hist')
```

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df,hue='lunch')
```

```python
sns.pairplot(data=df,hue='lunch',diag_kind='hist',corner=True)
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch',row='race/ethnicity')
```

```python
g=sns.PairGrid(df)
g=g.map_upper(sns.scatterplot)
g=g.map_diag(sns.boxplot)
g=g.map_lower(sns.kdeplot)
```
