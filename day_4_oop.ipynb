```python
# creating a class
class car:
    x = 3
    y = 5
r1 = car()
r2 = car()
print(r1.y)
print(r2.x)
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
p1 = person('faisal',23)
p2 = person('fatima','19')
print(p1.name,p1.age,sep = ' ')
print(p2.name,p2.age,sep = ' ')
class car:
    def __init__(self,brand,model,year):
        self.brand = brand
        self.model = model
        self.year = year
    def __str__(self):
        return f'{self.brand} {self.model} {self.year}'
    def mycar(self):
        print(f'my car is {self.brand} {self.model} {self.year}')
car1 = car('toyota','camry',2022)
print(car1)
car1.mycar()
class phone:
    def __init__(self,brand,year):
        self.brand = brand
        self.year = year
    def call(self):
        return f'{self.brand} is making a call'
phone1 = phone('samsung',2025)
phone2 = phone('techno',2026)
print(phone1.brand)
print(phone2.brand)
phone1.call()
phone2.call()
```

```python
class Vehicle:
    def __init__(self,brand,color):
        self.brand = brand
        self.color = color
    def start(self):
        return f'{self.brand} is starting'

    
class Car(Vehicle):
    def __init__(self,brand,color,year):
        super().__init__(brand,color)
        self.year = year
car1 = Car('toyota','white',200)
        
class plane(Vehicle):
    def __init__(self,brand,color,type):
        super().__init__(brand,color)
        self.type = type
plane1 = plane('BOENG','blue','f35')
print(plane1.start())
print(car1.start())
```

```python
class Electronics:
    def __init__(self,brand,power):
        self.brand = brand
        self.power = power
    
    def turn_on(self):
        return f'The device is powered on.'
    
class Phone(Electronics):
    def __init__(self,brand,power,sim_type):
        super().__init__(brand,power)
        self.sim_type = sim_type
    def make_call(self):
        return f'calling...'
phone = Phone('samsung','battery','dual sim')
    
class Smartphone(Phone):
    def __init__(self,brand,power,sim_type,os):
        super().__init__(brand,power,sim_type)
        self.os = os
    def install_app(self):
        return f'installing app on {self.os}'
smartphone = Smartphone('samsung','battery','dual sim','Android')
print('brand:',phone.brand)
print('power:',phone.power)
print('sim_type:',phone.sim_type)
print('os:',smartphone.os)
print(phone.turn_on())
print(phone.make_call())
print(smartphone.install_app())

```
